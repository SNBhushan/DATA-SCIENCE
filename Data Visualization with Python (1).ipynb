```python
#plotting multiple lines
years=range(2000,2012)
apples=[1,6,5,9,7,3,8,2,9,5,4,1]
oranges=[1,6,8,8,8,4,5,5,7,5,1,1]

plt.plot(years,apples)
plt.plot(years,oranges)

plt.xlabel('years')
plt.ylabel('tons(per hectre)');
```

```python
#Chart title and Legend
years=range(2000,2012)
apples=[1,6,5,9,7,3,8,2,9,5,4,1]
oranges=[1,6,8,8,8,4,5,5,7,5,1,5]

plt.plot(years,apples)
plt.plot(years,oranges)

plt.xlabel('years')
plt.ylabel('tons(per hectre)');

plt.title('yields per year')
plt.legend(['apples','oranges']);
```

```python
#line markers
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')
```

```python
#fmt argument
plt.plot(years,apples,'s--g')
plt.plot(years,oranges,'o-r')
```

```python
#if line style is not given only markers are drawn
plt.plot(years,apples,'sg')
plt.plot(years,oranges,'or');
```

```python
#changing the figure size
plt.figure(figsize=(12,6))
plt.plot(years,apples,'s--g')
plt.plot(years,oranges,'o-r');
plt.legend(['apples','oranges'])
```

# Histogtram 

```python
#vizualizing data of specific column
flowers_df.sepal_width
```

```python
plt.title('Distribution of sepal width')
plt.hist(flowers_df.sepal_width);
```

```python
#we can introduce bins
plt.title('Distribution of sepal width')
plt.hist(flowers_df.sepal_width, bins=15);
```

```python
import numpy as np
```

```python
plt.figure(figsize=(12,6))
plt.title('Distribution of sepal width')
plt.hist(flowers_df.sepal_width, bins=np.arange(2,5,0.25));
```

```python
plt.title('Distribution of sepal width')
plt.hist(flowers_df.sepal_width, bins=[2,2,3,4,5]);
```

```python
#Multiple Hoistogrms
setosa_df=flowers_df[flowers_df.species=="setosa"]
versi_df=flowers_df[flowers_df.species=="versicolor"]
verginica_df=flowers_df[flowers_df.species=="verginica"]
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,.25))
plt.hist(versi_df.sepal_width,alpha=0.4,bins=np.arange(2,5,.25))
plt.hist(verginica_df.sepal_width,alpha=0.4,bins=np.arange(2,5,.25))
plt.legend(['setosa','verginica','versicolor'])
```

```python
#we can stack histogrms on one another we need to give STACK=True
plt.figure(figsize=(12,6))
plt.hist([setosa_df.sepal_width,versi_df.sepal_width,verginica_df.sepal_width],bins=np.arange(2,5,.25),stacked=True);
plt.legend(['setosa','versicolor','verginica'])
```

# Bar chart

```python
years=range(2000,2006)
apples=[1,2,6,5,8,9]
oranges=[2,5,9,4,8,6]
plt.bar(years,apples)
plt.bar(years,oranges)
```

```python
#Can be stacked over one another by assining bottom="variable name"
plt.bar(years,apples)
plt.bar(years,oranges,bottom=apples);
```
